# Simulación del Truco

Las **cartas** deben registra el numero, palo y valor de la carta. 
El valor debe reflejar la jerarquía de las cartas en el juego del truco.

```python
def calcular_valor(numero, palo):
    jerarquia = {
        (1, "espada"): 14,
        (1, "basto"): 13,
        (7, "espada"): 12,
        (7, "oro"): 11,
        (3, "espada"): 10, (3, "basto"): 10, (3, "copa"): 10, (3, "oro"): 10,
        (2, "espada"): 9, (2, "basto"): 9, (2, "copa"): 9, (2, "oro"): 9,
        (12, "espada"): 8, (12, "basto"): 8, (12, "copa"): 8, (12, "oro"): 8,
        (11, "espada"): 7, (11, "basto"): 7, (11, "copa"): 7, (11, "oro"): 7,
        (10, "espada"): 6, (10, "basto"): 6, (10, "copa"): 6, (10, "oro"): 6,
        (7, "basto"): 5, (7, "copa"): 5,
        (6, "espada"): 4, (6, "basto"): 4, (6, "copa"): 4, (6, "oro"): 4,
        (5, "espada"): 3, (5, "basto"): 3, (5, "copa"): 3, (5, "oro"): 3,
        (4, "espada"): 2, (4, "basto"): 2, (4, "copa"): 2, (4, "oro"): 2,
    }
    return jerarquia.get((numero, palo), 1)
class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcular_valor(numero, palo)
    def __repr__(self):
        return f"{self.numero} de {self.palo} (Valor: {self.valor})"

c = Carta(3, "oro")
print(c)
```

**Mazo**: Representan las cartas españolas de 1 al 12, en los palos de oro, copa, espada y basto.

Debe existir una función que permita barajar las cartas.
y otra función que permita repartir las cartas a los jugadores.

```python
import random

class Mazo:
    def __init__(self):
        self.cartas = [Carta(numero, palo) for numero in range(1, 13) if numero != 8 and numero != 9
                       for palo in ["oro", "copa", "espada", "basto"]]
    def mezclar(self):
        random.shuffle(self.cartas)
    def repartir(self, cantidad):
        mano = self.cartas[:cantidad]
        self.cartas = self.cartas[cantidad:]
        return mano

m = Mazo()
m.mezclar()
print(m.repartir(3))
print(m.repartir(3))
```

**Jugador**: Deben tener un nombre y las cartas que tiene en la mano. Debe acumular los puntos ganados.

Nota: Para simplificar el problema, el jugador elige la carta a jugar de forma aleatoria.

```python
class Jugador:
    pass
    def __init__(self, nombre):
        self.nombre = nombre
        self.mano = []
        self.puntos = 0
    def recibir_cartas(self, cartas):
        self.mano.extend(cartas)
    def jugar_carta(self):
        if self.mano:
            carta_jugada = random.choice(self.mano)
            self.mano.remove(carta_jugada)
            return carta_jugada
        else:
            return None
    def sumar_puntos(self, puntos):
        self.puntos += puntos

j = Jugador("Juan")
m = Mazo()
m.mezclar()
j.recibir_cartas(m.repartir(3))
print(f"cartas en mano: {j.mano}")
print(f"carta jugada: {j.jugar_carta()}")
```

**Partida**: Debe tener dos jugadores, un mazo de cartas y un puntaje. Debe permitir jugar una carta por ronda y definir el ganador de la ronda. Debe permitir jugar hasta que se terminen las cartas del mazo.

```python
import random

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcular_valor(numero, palo)
    
    def __repr__(self):
        return f"{self.numero} de {self.palo} (Valor: {self.valor})"

def calcular_valor(numero, palo):
    jerarquia = {
        (1, "espada"): 14, (1, "basto"): 13,
        (7, "espada"): 12, (7, "oro"): 11,
        (3, "espada"): 10, (3, "basto"): 10, (3, "copa"): 10, (3, "oro"): 10,
        (2, "espada"): 9, (2, "basto"): 9, (2, "copa"): 9, (2, "oro"): 9,
        (12, "espada"): 8, (12, "basto"): 8, (12, "copa"): 8, (12, "oro"): 8,
        (11, "espada"): 7, (11, "basto"): 7, (11, "copa"): 7, (11, "oro"): 7,
        (10, "espada"): 6, (10, "basto"): 6, (10, "copa"): 6, (10, "oro"): 6,
        (7, "basto"): 5, (7, "copa"): 5,
        (6, "espada"): 4, (6, "basto"): 4, (6, "copa"): 4, (6, "oro"): 4,
        (5, "espada"): 3, (5, "basto"): 3, (5, "copa"): 3, (5, "oro"): 3,
        (4, "espada"): 2, (4, "basto"): 2, (4, "copa"): 2, (4, "oro"): 2,
    }
    return jerarquia.get((numero, palo), 1)

class Mazo:
    def __init__(self):
        self.cartas = [Carta(numero, palo) for numero in range(1, 13) if numero != 8 and numero != 9
                       for palo in ["oro", "copa", "espada", "basto"]]
    
    def mezclar(self):
        random.shuffle(self.cartas)
    
    def repartir(self, cantidad):
        mano = self.cartas[:cantidad]
        self.cartas = self.cartas[cantidad:]
        return mano

class Jugador:
    def __init__(self, nombre):
        self.nombre = nombre
        self.mano = []
        self.puntos = 0
    
    def recibir_cartas(self, cartas):
        self.mano = cartas
    
    def jugar_carta(self):
        if self.mano:
            carta_jugada = random.choice(self.mano)
            self.mano.remove(carta_jugada)
            return carta_jugada
        return None

    def sumar_puntos(self, puntos):
        self.puntos += puntos

class Partida:
    def __init__(self, jugador1, jugador2):
        self.jugador1 = jugador1
        self.jugador2 = jugador2
        self.mazo = Mazo()
        self.puntaje_maximo = 15

    def jugar_ronda(self, ronda_numero):
        print(f"\n--- Ronda {ronda_numero} ---")
        print(f"{self.jugador1.nombre} tiene las cartas: {self.jugador1.mano}")
        print(f"{self.jugador2.nombre} tiene las cartas: {self.jugador2.mano}")

        carta_jugador1 = self.jugador1.jugar_carta()
        carta_jugador2 = self.jugador2.jugar_carta()

        if carta_jugador1 and carta_jugador2:
            print(f"{self.jugador1.nombre} juega {carta_jugador1}")
            print(f"{self.jugador2.nombre} juega {carta_jugador2}")
            
            if carta_jugador1.valor > carta_jugador2.valor:
                print(f"Gana la ronda {self.jugador1.nombre}")
                self.jugador1.sumar_puntos(1)
            elif carta_jugador2.valor > carta_jugador1.valor:
                print(f"Gana la ronda {self.jugador2.nombre}")
                self.jugador2.sumar_puntos(1)
            else:
                print("La ronda termina en empate")
        else:
            print("No hay cartas suficientes para jugar la ronda.")

    def jugar(self):
        ronda_numero = 1
        while self.jugador1.puntos < self.puntaje_maximo and self.jugador2.puntos < self.puntaje_maximo:
            self.mazo.mezclar()

            self.jugador1.recibir_cartas(self.mazo.repartir(3))
            self.jugador2.recibir_cartas(self.mazo.repartir(3))
            
            for _ in range(3):
                self.jugar_ronda(ronda_numero)
                ronda_numero += 1

            print(f"\nPuntaje {self.jugador1.nombre}: {self.jugador1.puntos}")
            print(f"Puntaje {self.jugador2.nombre}: {self.jugador2.puntos}")

            if len(self.mazo.cartas) < 6:
                print("No quedan suficientes cartas en el mazo. Mezclando nuevamente.")
                self.mazo = Mazo()

        if self.jugador1.puntos >= self.puntaje_maximo:
            print(f"\n{self.jugador1.nombre} gana la partida!")
        elif self.jugador2.puntos >= self.puntaje_maximo:
            print(f"\n{self.jugador2.nombre} gana la partida!")

p = Partida(Jugador("Juan"), Jugador("Pedro"))
p.jugar()
```
