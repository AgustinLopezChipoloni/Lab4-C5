**Jugador**: Deben tener un nombre y las cartas que tiene en la mano. Debe acumular los puntos ganados.

Nota: Para simplificar el problema, el jugador elige la carta a jugar de forma aleatoria.

```python
import random

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcular_valor(numero, palo)

    def __repr__(self):
        return f'{self.numero} de {self.palo}'

def calcular_valor(numero, palo):
    jerarquia = {
        (1, 'espada'): 14,
        (1, 'basto'): 13,
        (7, 'espada'): 12,
        (7, 'oro'): 11,
        (3, 'espada'): 10, (3, 'basto'): 10, (3, 'oro'): 10, (3, 'copa'): 10,
        (2, 'espada'): 9, (2, 'basto'): 9, (2, 'oro'): 9, (2, 'copa'): 9,
        (1, 'oro'): 8, (1, 'copa'): 8,
        (12, 'espada'): 7, (12, 'basto'): 7, (12, 'oro'): 7, (12, 'copa'): 7,
        (11, 'espada'): 6, (11, 'basto'): 6, (11, 'oro'): 6, (11, 'copa'): 6,
        (10, 'espada'): 5, (10, 'basto'): 5, (10, 'oro'): 5, (10, 'copa'): 5,
        (7, 'basto'): 4, (7, 'copa'): 4,
        (6, 'espada'): 3, (6, 'basto'): 3, (6, 'oro'): 3, (6, 'copa'): 3,
        (5, 'espada'): 2, (5, 'basto'): 2, (5, 'oro'): 2, (5, 'copa'): 2,
        (4, 'espada'): 1, (4, 'basto'): 1, (4, 'oro'): 1, (4, 'copa'): 1
    }
    return jerarquia.get((numero, palo), 0)

class Mazo:
    def __init__(self):
        palos = ['oro', 'copa', 'espada', 'basto']
        numeros = range(1, 13)  
        self.cartas = [Carta(numero, palo) for palo in palos for numero in numeros]
    
    def mezclar(self):
        random.shuffle(self.cartas)

    def repartir(self, cantidad):
        if cantidad > len(self.cartas):
            return "No hay suficientes cartas en el mazo"
        mano = self.cartas[:cantidad]
        self.cartas = self.cartas[cantidad:]  
        return mano

class Jugador:
    def __init__(self, nombre):
        self.nombre = nombre
        self.mano = []  
        self.puntos = 0  
    
    def recibir_cartas(self, cartas):
        self.mano.extend(cartas)
    
    def jugar_carta(self):
        if self.mano:
            carta_a_jugar = random.choice(self.mano)
            self.mano.remove(carta_a_jugar)
            return carta_a_jugar
        else:
            return "No hay cartas para jugar"
    
    def ganar_puntos(self, puntos):

        self.puntos += puntos

    def __repr__(self):
        return f'Jugador({self.nombre}, Puntos: {self.puntos}, Cartas en mano: {len(self.mano)})'

m = Mazo()
m.mezclar()
j = Jugador("Juan")

j.recibir_cartas(m.repartir(3))

print(f"{j.nombre} juega la carta: {j.jugar_carta()}")


print(j)
```

**Partida**: Debe tener dos jugadores, un mazo de cartas y un puntaje. Debe permitir jugar una carta por ronda y definir el ganador de la ronda. Debe permitir jugar hasta que se terminen las cartas del mazo.

```python
import random

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = self.calcular_valor()

    def calcular_valor(self):
        jerarquia = {
            (1, 'espada'): 14,
            (1, 'basto'): 13,
            (7, 'espada'): 12,
            (7, 'oro'): 11,
            (3, 'espada'): 10, (3, 'basto'): 10, (3, 'oro'): 10, (3, 'copa'): 10,
            (2, 'espada'): 9, (2, 'basto'): 9, (2, 'oro'): 9, (2, 'copa'): 9,
            (1, 'oro'): 8, (1, 'copa'): 8,
            (12, 'espada'): 7, (12, 'basto'): 7, (12, 'oro'): 7, (12, 'copa'): 7,
            (11, 'espada'): 6, (11, 'basto'): 6, (11, 'oro'): 6, (11, 'copa'): 6,
            (10, 'espada'): 5, (10, 'basto'): 5, (10, 'oro'): 5, (10, 'copa'): 5,
            (7, 'basto'): 4, (7, 'copa'): 4,
            (6, 'espada'): 3, (6, 'basto'): 3, (6, 'oro'): 3, (6, 'copa'): 3,
            (5, 'espada'): 2, (5, 'basto'): 2, (5, 'oro'): 2, (5, 'copa'): 2,
            (4, 'espada'): 1, (4, 'basto'): 1, (4, 'oro'): 1, (4, 'copa'): 1
        }
        return jerarquia.get((self.numero, self.palo), 0)

    def __repr__(self):
        return f'{self.numero} de {self.palo} (Valor: {self.valor})'

class Mazo:
    def __init__(self):
        palos = ['oro', 'copa', 'espada', 'basto']
        numeros = range(1, 13)
        self.cartas = [Carta(numero, palo) for palo in palos for numero in numeros]
    
    def mezclar(self):
        random.shuffle(self.cartas)

    def repartir(self, cantidad):
        if len(self.cartas) < cantidad:
            return []
        mano = self.cartas[:cantidad]
        self.cartas = self.cartas[cantidad:]
        return mano

class Jugador:
    def __init__(self, nombre):
        self.nombre = nombre
        self.mano = []
        self.puntos = 0
    
    def recibir_cartas(self, cartas):
        self.mano = cartas

    def jugar_carta(self):
        if len(self.mano) == 0:
            raise ValueError(f"{self.nombre} no tiene cartas para jugar.")
        return self.mano.pop(0) 
    
    def ganar_puntos(self):
        self.puntos += 1

    def __repr__(self):
        return f'Jugador({self.nombre}, Puntos: {self.puntos}, Cartas en mano: {len(self.mano)})'

class Partida:
    def __init__(self, jugador1, jugador2):
        self.jugador1 = jugador1
        self.jugador2 = jugador2
        self.mazo = Mazo()

    def jugar_ronda(self):
        carta1 = self.jugador1.jugar_carta()
        carta2 = self.jugador2.jugar_carta()
        
        print(f"{self.jugador1.nombre} juega {carta1}")
        print(f"{self.jugador2.nombre} juega {carta2}")
        
        if carta1.valor > carta2.valor:
            print(f"Gana la ronda {self.jugador1.nombre}")
            self.jugador1.ganar_puntos()
        elif carta2.valor > carta1.valor:
            print(f"Gana la ronda {self.jugador2.nombre}")
            self.jugador2.ganar_puntos()
        else:
            print("La ronda es un empate")
    
    def jugar(self):
        self.mazo.mezclar()

        while self.jugador1.puntos < 15 and self.jugador2.puntos < 15:
            print(f"\nIniciando nueva partida... Puntaje actual: {self.jugador1.nombre}: {self.jugador1.puntos}, {self.jugador2.nombre}: {self.jugador2.puntos}")
            
            if len(self.mazo.cartas) < 6:  
                print("No hay suficientes cartas para repartir. El juego termina.")
                break
            
            self.jugador1.recibir_cartas(self.mazo.repartir(3))
            self.jugador2.recibir_cartas(self.mazo.repartir(3))

            for _ in range(3):
                if len(self.jugador1.mano) == 0 or len(self.jugador2.mano) == 0:
                    print("No hay más cartas para jugar")
                    break
                self.jugar_ronda()

        if self.jugador1.puntos >= 15:
            print(f"\n¡{self.jugador1.nombre} gana el juego con {self.jugador1.puntos} puntos!")
        elif self.jugador2.puntos >= 15:
            print(f"\n¡{self.jugador2.nombre} gana el juego con {self.jugador2.puntos} puntos!")

p = Partida(Jugador("Juan"), Jugador("Pedro"))
p.jugar()
```
