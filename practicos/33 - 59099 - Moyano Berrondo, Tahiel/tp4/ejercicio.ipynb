# Simulación del Truco

Las **cartas** deben registra el numero, palo y valor de la carta. 
El valor debe reflejar la jerarquía de las cartas en el juego del truco.

```python
def calcular_valor(numero, palo):
    #palos y valores
    jerarquia = {
        (1, 'espadas'): 14, 
        (1, 'bastos'): 13, 
        (7, 'espadas'): 12, 
        (7, 'oro'): 11, 
        (3, 'espadas'): 10,
        (3, 'bastos'): 10,
        (3, 'oro'): 10, 
        (3, 'copas'): 10,
        (2, 'espadas'): 9, 
        (2, 'bastos'): 9, 
        (2, 'oro'): 9, 
        (2, 'copas'): 9,
        (1, 'copas'): 8, 
        (1, 'oro'): 8,
        (12, 'espadas'): 7, 
        (12, 'bastos'): 7, 
        (12, 'oro'): 7, 
        (12, 'copas'): 7,
        (11, 'espadas'): 6, 
        (11, 'bastos'): 6, 
        (11, 'oro'): 6, 
        (11, 'copas'): 6,
        (10, 'espadas'): 5, 
        (10, 'bastos'): 5, 
        (10, 'oro'): 5, 
        (10, 'copas'): 5,
        (7, 'bastos'): 4, 
        (7, 'copas'): 4,
        (6, 'espadas'): 3, 
        (6, 'bastos'): 3, 
        (6, 'oro'): 3, 
        (6, 'copas'): 3,
        (5, 'espadas'): 2,
        (5, 'bastos'): 2,
        (5, 'oro'): 2, 
        (5, 'copas'): 2,
        (4, 'espadas'): 1, 
        (4, 'bastos'): 1, 
        (4, 'oro'): 1, 
        (4, 'copas'): 1
    }

    return jerarquia.get((numero, palo), 0)
#representa las cartas
class Carta:
    def __init__(self, numero, palo):
        self.numero=numero
        self.palo=palo
        self.valor=calcular_valor(numero, palo)
    def __str__(self) -> str:
        return f"{self.numero} de {self.palo} (valor: {self.valor})" 

palos=['espadas', 'bastos', 'oro', 'copas']
numeros=[1,2,3,4,5,6,7,10,11,12]

mazo=[Carta(numero, palo)for palo in palos for numero in numeros]

for Carta in mazo:
    print(Carta)

```

**Mazo**: Representan las cartas españolas de 1 al 12, en los palos de oro, copa, espada y basto.

Debe existir una función que permita barajar las cartas.
y otra función que permita repartir las cartas a los jugadores.

```python
import random

class Mazo:
    def __init__(self):
        self.palos = ['espadas', 'bastos', 'oro', 'copas']
        self.numeros=[1,2,3,4,5,6,7,10,11,12]
        self.cartas=[(numero, palo) for palo in self.palos for numero in self.numeros]
    def mezclar(self):
        random.shuffle(self.cartas)
    def repartir(self, cantidad):
        if cantidad > len(self.cartas):
            raise ValueError("No hay cartas en el mazo")
        mano=self.cartas[:cantidad]
        self.cartas = self.cartas[cantidad:]
        return mano

m = Mazo()
m.mezclar()
print(m.repartir(3))
```

**Jugador**: Deben tener un nombre y las cartas que tiene en la mano. Debe acumular los puntos ganados.

Nota: Para simplificar el problema, el jugador elige la carta a jugar de forma aleatoria.

```python

class Jugador:
    
    def __init__(self, nombre):
        self.nombre = nombre
        self.cartas = []
        self.puntos = 0
    
    def recibir_cartas(self, cartas):
        self.cartas = cartas 
    
    def jugar_carta(self):
        if self.cartas:  
            jugada = self.cartas[0]  
            self.cartas.remove(jugada) 
            return jugada
        return None

# Ejemplo de uso
m = Mazo()  
m.mezclar() 

j = Jugador("Messi") 
j.recibir_cartas(m.repartir(3))  # Reparte 3 cartas


print(f"{j.nombre} juega la carta: {j.jugar_carta()}")
```

**Partida**: Debe tener dos jugadores, un mazo de cartas y un puntaje. Debe permitir jugar una carta por ronda y definir el ganador de la ronda. Debe permitir jugar hasta que se terminen las cartas del mazo.

```python
import random

class Jugador:
    def __init__(self, nombre):
        self.nombre = nombre
        self.cartas = [] 
        self.puntos = 0
    
    def recibir_cartas(self, cartas):
        self.cartas = cartas 
    
    def jugar_carta(self):
        if self.cartas:  
            carta_jugada = self.cartas[0]  
            self.cartas.remove(carta_jugada)  
            return carta_jugada
        return None

class Partida:
    def __init__(self, jugador1, jugador2):
        self.jugador1 = jugador1
        self.jugador2 = jugador2
        self.cartas = [(numero, palo) for palo in ['espadas', 'bastos', 'oro', 'copas']
                       for numero in range(1, 14)]  
        self.puntaje_ganador = 15 

    def jugar_ronda(self):
       
        carta_jugador1 = self.jugador1.jugar_carta()
        carta_jugador2 = self.jugador2.jugar_carta()

        if carta_jugador1 is None or carta_jugador2 is None:
            print("No hay suficientes cartas para jugar.")
            return
        
        print(f"{self.jugador1.nombre} juega: {carta_jugador1[0]} de {carta_jugador1[1]}")
        print(f"{self.jugador2.nombre} juega: {carta_jugador2[0]} de {carta_jugador2[1]}")
        
        if carta_jugador1[0] > carta_jugador2[0]:  
            print(f"{self.jugador1.nombre} gana la ronda.\n")
            self.jugador1.puntos += 1
        elif carta_jugador1[0] < carta_jugador2[0]:
            print(f"{self.jugador2.nombre} gana la ronda.\n")
            self.jugador2.puntos += 1
        else:
            print("La ronda es un empate.\n")

    def jugar(self):
        while self.jugador1.puntos < self.puntaje_ganador and self.jugador2.puntos < self.puntaje_ganador:
            random.shuffle(self.cartas)  # Mezclar el mazo
            # Repartir 3 cartas a cada jugador
            self.jugador1.recibir_cartas(self.cartas[:3])
            self.jugador2.recibir_cartas(self.cartas[3:6])
            print(f"{self.jugador1.nombre} tiene: {self.jugador1.cartas}")
            print(f"{self.jugador2.nombre} tiene: {self.jugador2.cartas}")

            # Jugar una ronda
            self.jugar_ronda()

        # Determinar el ganador de la partida
        if self.jugador1.puntos >= self.puntaje_ganador:
            print(f"{self.jugador1.nombre} gana la partida con {self.jugador1.puntos} puntos.")
        else:
            print(f"{self.jugador2.nombre} gana la partida con {self.jugador2.puntos} puntos.")

# Ejemplo de uso
p = Partida(Jugador("Messi"), Jugador("Licha"))
p.jugar()
```
