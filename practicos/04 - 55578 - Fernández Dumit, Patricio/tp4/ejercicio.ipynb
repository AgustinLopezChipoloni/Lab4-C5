# Simulación del Truco

Las **cartas** deben registra el numero, palo y valor de la carta. 
El valor debe reflejar la jerarquía de las cartas en el juego del truco.

```python
def calcular_valor(numero, palo):
    
    valores_truco = {
        "espada": {1: 14, 7: 12, 3: 10, 2: 9, 12: 8, 11: 7, 10: 6, 6: 4, 5: 3, 4: 2},
        "basto": {1: 13, 7: 5, 3: 10, 2: 9, 12: 8, 11: 7, 10: 6, 6: 4, 5: 3, 4: 2},
        "oro": {3: 10, 2: 9, 1: 8, 12: 7, 11: 6, 10: 5, 7: 11, 6: 4, 5: 3, 4: 2},
        "copa": {3: 10, 2: 9, 1: 8, 12: 7, 11: 6, 10: 6, 7: 5, 6: 4, 5: 3, 4: 2},
    }
    
    return valores_truco.get(palo, {}).get(numero, 0)

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcular_valor(numero, palo)

    def __str__(self):
        return f"{self.numero} de {self.palo} (Valor: {self.valor})"

c = Carta(1, "espada")
print(c)  
```

**Mazo**: Representan las cartas españolas de 1 al 12, en los palos de oro, copa, espada y basto.

Debe existir una función que permita barajar las cartas.
y otra función que permita repartir las cartas a los jugadores.

```python
import random

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcular_valor(numero, palo)

    def __str__(self):
        return f"{self.numero} de {self.palo} (Valor: {self.valor})"

def calcular_valor(numero, palo):
    valores_truco = {
        "espada": {1: 14, 7: 12, 3: 10, 2: 9, 12: 8, 11: 7, 10: 6, 6: 4, 5: 3, 4: 2},
        "basto": {1: 13, 7: 5, 3: 10, 2: 9, 12: 8, 11: 7, 10: 6, 6: 4, 5: 3, 4: 2},
        "oro": {3: 10, 2: 9, 1: 8, 12: 7, 11: 6, 10: 5, 7: 11, 6: 4, 5: 3, 4: 2},
        "copa": {3: 10, 2: 9, 1: 8, 12: 7, 11: 6, 10: 6, 7: 5, 6: 4, 5: 3, 4: 2},
    }
    
    return valores_truco.get(palo, {}).get(numero, 0)

class Mazo:
    def __init__(self):
        self.cartas = self.crear_mazo()

    def crear_mazo(self):
        palos = ["oro", "copa", "espada", "basto"]
        numeros = [1, 2, 3, 4, 5, 6, 7, 10, 11, 12]  
        mazo = []

        for palo in palos:
            for numero in numeros:
                mazo.append(Carta(numero, palo))

        return mazo

    def mezclar(self):
        random.shuffle(self.cartas)

    def repartir(self, cantidad):
        if cantidad > len(self.cartas):
            raise ValueError("No hay suficientes cartas en el mazo para repartir")
        
        cartas_repartidas = self.cartas[:cantidad]
        self.cartas = self.cartas[cantidad:]
        return cartas_repartidas

m = Mazo()
m.mezclar()
cartas_repartidas = m.repartir(3)
for carta in cartas_repartidas:
    print(carta)  
```

**Jugador**: Deben tener un nombre y las cartas que tiene en la mano. Debe acumular los puntos ganados.

Nota: Para simplificar el problema, el jugador elige la carta a jugar de forma aleatoria.

```python
import random

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcular_valor(numero, palo)

    def __str__(self):
        return f"{self.numero} de {self.palo} (Valor: {self.valor})"

def calcular_valor(numero, palo):
    valores_truco = {
        "espada": {1: 14, 7: 12, 3: 10, 2: 9, 12: 8, 11: 7, 10: 6, 6: 4, 5: 3, 4: 2},
        "basto": {1: 13, 7: 5, 3: 10, 2: 9, 12: 8, 11: 7, 10: 6, 6: 4, 5: 3, 4: 2},
        "oro": {3: 10, 2: 9, 1: 8, 12: 7, 11: 6, 10: 5, 7: 11, 6: 4, 5: 3, 4: 2},
        "copa": {3: 10, 2: 9, 1: 8, 12: 7, 11: 6, 10: 6, 7: 5, 6: 4, 5: 3, 4: 2},
    }
    
    return valores_truco.get(palo, {}).get(numero, 0)

class Mazo:
    def __init__(self):
        self.cartas = self.crear_mazo()

    def crear_mazo(self):
        palos = ["oro", "copa", "espada", "basto"]
        numeros = [1, 2, 3, 4, 5, 6, 7, 10, 11, 12] 
        mazo = []

        for palo in palos:
            for numero in numeros:
                mazo.append(Carta(numero, palo))

        return mazo

    def mezclar(self):
        random.shuffle(self.cartas)

    def repartir(self, cantidad):
        if cantidad > len(self.cartas):
            raise ValueError("No hay suficientes cartas en el mazo para repartir")
        
        cartas_repartidas = self.cartas[:cantidad]
        self.cartas = self.cartas[cantidad:] 
        return cartas_repartidas

class Jugador:
    def __init__(self, nombre):
        self.nombre = nombre
        self.cartas = []     
        self.puntos = 0   

    def recibir_cartas(self, cartas):
        self.cartas.extend(cartas)   

    def jugar_carta(self):
        if not self.cartas:
            return None  
        
        carta_jugada = random.choice(self.cartas)   
        self.cartas.remove(carta_jugada)          
        return carta_jugada
      
m = Mazo()
m.mezclar()
j = Jugador("Juan")
j.recibir_cartas(m.repartir(3))

print(f"Cartas en mano de {j.nombre}:")
for carta in j.cartas:
    print(carta)
    
carta_jugada = j.jugar_carta()
print(f"{j.nombre} juega la carta: {carta_jugada}")
```

**Partida**: Debe tener dos jugadores, un mazo de cartas y un puntaje. Debe permitir jugar una carta por ronda y definir el ganador de la ronda. Debe permitir jugar hasta que se terminen las cartas del mazo.

```python
import random

class Jugador:
    def __init__(self, nombre):
        self.nombre = nombre
        self.cartas = []
        self.puntos = 0

    def jugar_carta(self, indice):
        return self.cartas.pop(indice)

    def tiene_cartas(self):
        return len(self.cartas) > 0

class Partida: 
    def __init__(self, jugador1, jugador2):
        self.jugador1 = jugador1
        self.jugador2 = jugador2
        self.mazo = self.crear_mazo()

    def crear_mazo(self):
        palos = ["oro", "copa", "espada", "basto"]
        return [Carta(numero, palo) for palo in palos for numero in range(1, 13) if numero not in (8, 9)]

    def mezclar_y_repartir(self):
        random.shuffle(self.mazo)
        self.jugador1.cartas = self.mazo[:3]
        self.jugador2.cartas = self.mazo[3:6]

    def jugar_ronda(self):
        if not self.jugador1.tiene_cartas() or not self.jugador2.tiene_cartas():
            print("Repartiendo cartas ...")
            self.mezclar_y_repartir()

        carta1 = self.jugador1.jugar_carta(0)
        carta2 = self.jugador2.jugar_carta(0)

        print(f"{self.jugador1.nombre} juega: {carta1}")
        print(f"{self.jugador2.nombre} juega: {carta2}")

        if carta1.valor > carta2.valor:
            self.jugador1.puntos += 1
            print(f"{self.jugador1.nombre} gana la ronda.")
        elif carta2.valor > carta1.valor:
            self.jugador2.puntos += 1
            print(f"{self.jugador2.nombre} gana la ronda.")
        else:
            print("La ronda termina en empate.")

    def jugar(self):
        self.mezclar_y_repartir()

        while self.jugador1.puntos < 15 and self.jugador2.puntos < 15:
            self.jugar_ronda()

        if self.jugador1.puntos >= 15:
            print(f"¡{self.jugador1.nombre} gana la partida!")
        else:
            print(f"¡{self.jugador2.nombre} gana la partida!")

p = Partida(Jugador("Juan"), Jugador("Pedro"))
p.jugar()
```
