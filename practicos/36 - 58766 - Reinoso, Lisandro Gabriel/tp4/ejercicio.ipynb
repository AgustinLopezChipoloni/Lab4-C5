Las **cartas** deben registra el numero, palo y valor de la carta. 
El valor debe reflejar la jerarqu칤a de las cartas en el juego del truco.

```python
def calcular_valor(numero, palo):
    
    valores = {
        (1, "espada"): 14,
        (1, "basto"): 13,
        (7, "espada"): 12,
        (7, "oro"): 11,
        (3,"espada"): 10,
        (3,"oro"): 10,
        (3,"basto"): 10,
        (3,"copa"): 10,
        (2,"basto"): 9,
        (2,"espada"): 9,
        (2,"copa"): 9,
        (2,"oro"): 9,
        (1,"copa"): 8,
        (1, "oro"): 8,
        (12,"oro"): 7,
        (12,"espada"): 7,
        (12,"copa"): 7,
        (12,"basto"): 7,
        (11,"copa"): 6,
        (11,"basto"): 6,
        (11,"espada"): 6,
        (11,"oro"): 6,
        (10,"oro"): 5,
        (10,"basto"): 5,
        (10,"copa"): 5,
        (10,"espada"): 5,
        (7,"copa"): 4,
        (7,"basto"): 4,
        (6,"espada"): 3,
        (6,"basto"): 3,
        (6,"oro"): 3,
        (6,"copa"): 3,
        (5,"oro"): 2,
        (5,"espada"): 2,
        (5,"copa"): 2,
        (5,"basto"): 2,
        (4,"basto"): 1,
        (4,"copa"): 1,
        (4,"espada"): 1,
        (4,"oro"): 1    
    }
    
    return valores.get((numero,palo))
        

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcular_valor(numero, palo)
        
    
    def __str__(self):
        return f"La carta con numero {self.numero} y palo {self.palo} vale {self.valor} puntos"
           

c = Carta(1,"oro")
print(c)
```

**Mazo**: Representan las cartas espa침olas de 1 al 12, en los palos de oro, copa, espada y basto.

Debe existir una funci칩n que permita barajar las cartas.
y otra funci칩n que permita repartir las cartas a los jugadores.

```python
import random
import itertools


numeros = [1,2,3,4,5,6,7,10,11,12]
palos = ["oro", "basto", "espada", "copa"]
mazo = list(itertools.product(numeros, palos))

     
class Mazo:
    def __init__(self):
        self.mazo = mazo
        
    def mezclar(self):
        random.shuffle(self.mazo)
        
    def repartir(self, cantidad):
        tres_cartas = self.mazo[:cantidad]
        self.mazo = self.mazo[cantidad:]
        return tres_cartas
    
    def __str__(self):
            return f"{self.mazo} ahora el mazo tiene {len(self.mazo)} cartas"

m = Mazo()
m.mezclar()
print(m.repartir(3))
```

**Jugador**: Deben tener un nombre y las cartas que tiene en la mano. Debe acumular los puntos ganados.

Nota: Para simplificar el problema, el jugador elige la carta a jugar de forma aleatoria.

```python
class Jugador:
    def __init__(self, nombre):
        self.nombre = nombre
        self.cartas_jugador = []
        self.puntos = 0
        
    def recibir_cartas(self, cartas):
        self.cartas_jugador = cartas
        return f"El jugador {self.nombre} tiene las cartas {self.cartas_jugador}"
    
    def jugar_carta(self):
        decision_random = random.randint(0, len(self.cartas_jugador)-1)
        carta_a_jugar = self.cartas_jugador.pop(decision_random)
        
        return carta_a_jugar
    
   
        


j = Jugador("Juan")
j.recibir_cartas(m.repartir(3))
print(j.jugar_carta())
```
