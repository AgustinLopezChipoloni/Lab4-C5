## Ejercicio 1: Cargar Datos de ventas.

El archivo datos.dat contiene el registro de las ventas realizadas.

Tiene un formato de ancho fijo:
- fecha:    10 lugares
- producto: 30 lugares
- precio:   10 lugares
- cantidad:  5 lugares

hacer una funcion que carge los en una lista de diccionarios. 

nota: con 
```python
lineas = open("datos.dat", "r").readlines() 
```
puede obtener la lista de string de todos los datos 

````python
def cargar_datos():
    ventas = []
    
    lineas = open("datos.dat", "r").readlines()
    
    for linea in lineas:
        fecha = linea[0:10].strip()     
        producto = linea[10:40].strip()  
        precio = linea[40:50].strip()   
        cantidad = linea[50:55].strip()   
        
        venta = {
            "fecha": fecha,
            "producto": producto,
            "precio": float(precio), 
            "cantidad": int(cantidad)  
        }
        
        ventas.append(venta)
    
    return ventas

datos = cargar_datos()

for venta in datos:
    print(venta)
````

## Ejercicio 2: Calcular el total de ventas.

Hacer una funcion que sume los importes vendidos (precio * cantidad) y las cantidades

````python
def cargar_datos():
    ventas = []
    
    lineas = open("datos.dat", "r").readlines()
    
    for linea in lineas:
        fecha = linea[0:10].strip()     
        producto = linea[10:40].strip() 
        precio = linea[40:50].strip()    
        cantidad = linea[50:55].strip() 
        
        venta = {
            "fecha": fecha,
            "producto": producto,
            "precio": float(precio),  
            "cantidad": int(cantidad)  
        }
        
        ventas.append(venta)
    
    return ventas

def calcular_totales(ventas):
    total_importe = 0.0
    total_cantidad = 0
    
    for venta in ventas:
        precio = venta["precio"]
        cantidad = venta["cantidad"]
        
        total_importe += precio * cantidad
        total_cantidad += cantidad
    
    return total_importe, total_cantidad

datos = cargar_datos()

importe, cantidad = calcular_totales(datos)

print(f"Las ventas fueron de ${importe:.2f} en {cantidad} unidades")
````

## Ejercicio 3: Listar las unidades vendidas.

Listar cuantas unidades se vendieron en total para cada producto

````python
def unidades_vendidas(datos):
    ventas_productos = {}

    for venta in datos:
        producto = venta['producto'] 
        cantidad = venta['cantidad']  

        if producto in ventas_productos:
            ventas_productos[producto] += cantidad
        else:
            ventas_productos[producto] = cantidad

    return ventas_productos  

def listar_ventas(ventas):
    for producto, cantidad in ventas.items():
        print(f"Producto: {producto}, unidades vendidas: {cantidad}")

def cargar_datos():
    lineas = open("datos.dat", "r").readlines()
    datos = []

    for linea in lineas:
        venta = {
            'fecha': linea[0:10].strip(),
            'producto': linea[10:40].strip(),
            'precio': float(linea[40:50].strip()),
            'cantidad': int(linea[50:55].strip())
        }
        datos.append(venta)
    
    return datos

datos = cargar_datos()

ventas = unidades_vendidas(datos)

listar_ventas(ventas)
````

##  Ejercicio 4: Listar el precio promedio por producto.

Hacer un listado del precio promedio por productos.

````python
def precio_promedio(datos):
    precios_totales = {}
    
    for venta in datos:
        producto = venta['producto']  
        precio = venta['precio']       
        cantidad = venta['cantidad']   

        if producto in precios_totales:
            precios_totales[producto]['suma_precios'] += precio * cantidad
            precios_totales[producto]['total_vendido'] += cantidad
        else:
            precios_totales[producto] = {
                'suma_precios': precio * cantidad,
                'total_vendido': cantidad
            }

    precios_promedio = {}
    for producto, datos in precios_totales.items():
        precios_promedio[producto] = datos['suma_precios'] / datos['total_vendido']

    return precios_promedio

def listar_precios(precios):
    for producto, precio in precios.items():
        print(f"Producto: {producto}, precio promedio: ${precio:.2f}")

def cargar_datos():
    lineas = open("datos.dat", "r").readlines()
    datos = []

    for linea in lineas:
        venta = {
            'fecha': linea[0:10].strip(),
            'producto': linea[10:40].strip(),
            'precio': float(linea[40:50].strip()),
            'cantidad': int(linea[50:55].strip())
        }
        datos.append(venta)
    
    return datos

datos = cargar_datos()

precios = precio_promedio(datos)

listar_precios(precios)
````

### Ejercicio 5: Ranking de productos

Realizar un listado de los 3 productos mas vendidos ordenados la cantidad de unidades vendidas (ordenadas de mayor a menor)


````python
def ranking_productos(datos, top=3):
    ventas_producto = {}

    for venta in datos:
        producto = venta['producto']  
        cantidad = venta['cantidad']  
        
        # Acumular la cantidad vendida por producto
        if producto in ventas_producto:
            ventas_producto[producto] += cantidad
        else:
            ventas_producto[producto] = cantidad

    ranking = sorted(ventas_producto.items(), key=lambda x: x[1], reverse=True)

    return ranking[:top]

def listar_ranking(ranking):
    print("Ranking de productos más vendidos:")
    for producto, cantidad in ranking:
        print(f"Producto: {producto}, Unidades vendidas: {cantidad}")

lineas = open("datos.dat", "r").readlines()

datos = []
for linea in lineas:
    fecha = linea[0:10].strip()
    producto = linea[10:40].strip()
    precio = float(linea[40:50].strip())
    cantidad = int(linea[50:55].strip())

    datos.append({
        "fecha": fecha,
        "producto": producto,
        "precio": precio,
        "cantidad": cantidad
    })

ranking = ranking_productos(datos)
listar_ranking(ranking)
````

### Ejercicio 6: Listar las ventas por mes

Realizar un listado del total de unidades vendidas por productos separada por mes.

````python
def ventas_por_mes(datos):
    ventas_mensuales = {}

    for venta in datos:
        fecha = venta['fecha']  
        producto = venta['producto'] 
        cantidad = venta['cantidad'] 

        mes = fecha[3:5]  
        año = fecha[6:10]  
        mes_año = f"{año}-{mes}"  

        if mes_año not in ventas_mensuales:
            ventas_mensuales[mes_año] = {}

        if producto not in ventas_mensuales[mes_año]:
            ventas_mensuales[mes_año][producto] = 0

        ventas_mensuales[mes_año][producto] += cantidad

    return ventas_mensuales

def listar_ventas_mensuales(ventas):
    for mes_año, productos in ventas.items():
        print(f"Ventas para el mes {mes_año}:")
        for producto, cantidad in productos.items():
            print(f"  Producto: {producto}, Unidades vendidas: {cantidad}")
        print("") 

lineas = open("datos.dat", "r").readlines()

datos = []
for linea in lineas:
    fecha = linea[0:10].strip()
    producto = linea[10:40].strip()
    precio = float(linea[40:50].strip())
    cantidad = int(linea[50:55].strip())

    datos.append({
        "fecha": fecha,
        "producto": producto,
        "precio": precio,
        "cantidad": cantidad
    })

ventas = ventas_por_mes(datos)
listar_ventas_mensuales(ventas)
````

### Ejercicio 7: Informe general

Mostrar un listado de productos ordenados alfabeticamente que contengan el precio promedio, la cantidad de unidades vendidas y el importe total vendido para cada producto

````python
def resumen_ventas(datos):
    resumen = {}
    
    for linea in datos:
        producto = linea[10:40].strip()  
        precio = float(linea[40:50].strip())  
        cantidad = int(linea[50:55].strip())  

        importe_total = precio * cantidad
        
        if producto in resumen:
            resumen[producto]['cantidad'] += cantidad
            resumen[producto]['importe_total'] += importe_total
        else:
            resumen[producto] = {
                'precio_promedio': precio,
                'cantidad': cantidad,
                'importe_total': importe_total
            }
    
    return resumen

def informe_ventas(datos):
    productos_ordenados = sorted(datos.keys())
    
    print("Listado de Productos:")
    print("---------------------")  
    
    for producto in productos_ordenados:
        info = datos[producto]
        print("Producto:", producto)
        print("  Precio Promedio:", info['precio_promedio'])
        print("  Unidades Vendidas:", info['cantidad'])
        print("  Importe Total:", info['importe_total'])
        print() 

lineas = open("datos.dat", "r").readlines()
resumen = resumen_ventas(lineas)
informe_ventas(resumen)
````
